```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Load the latest uploaded dataset
df = pd.read_csv("/content/CVD_FinalData.csv")
```

```python
# Set up the visual style
sns.set(style="whitegrid")
```

```python
# Summary statistics and null check
summary = df.describe()
missing_values = df.isnull().sum()
```

```python
# Visualization 1: Distribution of the target variable
plt.figure(figsize=(6, 4))
sns.countplot(x='TenYearCHD', data=df, palette='Set2')
plt.title('Distribution of 10-Year CHD Risk')
plt.xlabel('TenYearCHD (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Visualization 2: Correlation heatmap
plt.figure(figsize=(14, 10))
# Select only numerical features for correlation
numerical_df = df.select_dtypes(include=['number'])  # Select columns with numerical data types
corr = numerical_df.corr()  # Calculate correlation on the numerical subset
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
# Visualization 3: Age distribution by CHD risk
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df, x='age', hue='TenYearCHD', fill=True, common_norm=False, palette='crest')
plt.title('Age Distribution by CHD Risk')
plt.xlabel('Age')
plt.ylabel('Density')
plt.tight_layout()
plt.show()
```

```python
# Visualization 4: Boxplot of cholesterol levels by CHD risk
plt.figure(figsize=(8, 5))
sns.boxplot(x='TenYearCHD', y='totChol', data=df, palette='pastel')
plt.title('Total Cholesterol by CHD Risk')
plt.xlabel('TenYearCHD')
plt.ylabel('Total Cholesterol')
plt.tight_layout()
plt.show()
```
